```python
s = Service('D:\Selenium driver\chromedriver2.exe')
driver = webdriver.Chrome(service=s)
url = 'https://www.booking.com/searchresults.en-gb.html?label=gen173nr-1FCAEoggI46AdIM1gEaBSIAQGYAQm4ARfIAQzYAQHoAQH4AQuIAgGoAgO4AtGK3Y0GwAIB0gIkNTBkMDMyNTYtYjZjMi00ZDFhLTg5YmEtZjJkNjA1NjRjYTcy2AIG4AIB;sid=c265df03d9477e71793a9ca2000cbe3a;dest_id=211349;dest_type=city&'
driver.get(url)
time.sleep(2)
date_tab = driver.find_element(By.XPATH, '//div[@class="sb-date-field__display"]')
date_tab.click()
```

```python
list_of_links = []
condition = True
while condition:
    hotels = driver.find_elements_by_xpath('//div[@class="a5c3ef8268"]/a')
    for hotel in hotels:
        link = hotel.get_property('href')
        list_of_links.append(link)
    try:
        next = driver.find_element_by_xpath('//div[@class="ce83a38554 _ea2496c5b"]')
        next.click()
        time.sleep(2)
    except:
        condition = False
```

```python
for idx, link in enumerate(list_of_links):
    driver.get(link)
    hotel_name = driver.find_element_by_id('hp_hotel_name').text.replace('Hotel ', '')
    print(idx)
    review_button = driver.find_element_by_xpath('(//li[@class="hp_nav_bar_item"])[4]/a')
    review_button.click()
    time.sleep(2)
    with open('Booking_.csv', 'a', newline='', encoding="utf-8") as f:
        the_writer= writer(f)
        header = ['Hotel Name', 'Customer Name', 'Rating', 'Country', 'Comment']
        the_writer.writerow(header)
        condition = True

        while condition :
            reviews = driver.find_elements_by_xpath('//div[@class="c-review-block"]')
            for i in reviews:
                try:
                    name = i.find_element_by_class_name('bui-avatar-block__title').text
                except Exception:
                    name = 'None'
                try: 
                    rating = i.find_element_by_class_name('bui-review-score__badge').text
                except Exception:
                    rating = 'None'
                try:
                    country = i.find_element_by_class_name('bui-avatar-block__subtitle').text
                except Exception:
                    country = 'None'

                try:
                    comment = i.find_element_by_class_name('c-review__title--ltr').text
                except Exception:
                    comment = 'None'


                info = [hotel_name, name, rating, country, comment]
                the_writer.writerow(info)
                time.sleep(1)
                print(name)
                print(country)
                print(rating)
                print(comment)


            try:
                next = driver.find_element_by_xpath('//a[@class="pagenext"]')
                next.click()
                time.sleep(2)
            except :
                condition = False
```
