```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException, TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from csv import writer
import csv
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
s = Service('D:\Selenium driver\chromedriver2.exe')
driver = webdriver.Chrome(service=s)
url = 'https://www.airbnb.com/rooms/43871809/photos?guests=1&adults=1'
driver.get(url)
time.sleep(4)
```

```python
images = driver.find_elements_by_xpath('//div[@data-testid="photo-viewer-section"]//a')
```

```python
len(images)
    
```

```python

with open('airbnb_images.csv', 'w', encoding='utf-8', newline='') as f:
    the_writer = writer(f)
    header = ['Name', 'Image']
    the_writer.writerow(header)
    name = driver.find_element_by_xpath('//h1[@class="_fecoyn4"]').text
    for each in images:
        image_link = each.get_attribute('href')
        print(image_link)
        print(name)
        info = [name, image_link]
        the_writer.writerow(info)
```

```python

name = driver.find_element_by_xpath('//h1[@class="_fecoyn4"]').text
with open('image_link.csv', 'w', encoding='utf-8', newline='') as f:
    
    the_writer = writer(f)
    header = ['Name', 'Image_link']
    the_writer.writerow(header)
    for each in images:
        image_link = each.find_element_by_xpath('//img[@class="_6tbg2q"]')
        link = image_link.get_attribute('src')
        info = [link]
        the_writer.writerow(info)
        print(name)
        print(link)    
        print()
            
    
```
