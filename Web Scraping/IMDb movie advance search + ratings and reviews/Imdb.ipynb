```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from csv import writer
import requests
import csv
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
s = Service('D:\Selenium driver\chromedriver.exe')
driver = webdriver.Chrome(service=s)
driver.get('https://imdb.com')
```

```python
# dropdown
time.sleep(1)
all_button = driver.find_element(By.CLASS_NAME, 'ipc-icon--arrow-drop-down')
all_button.click()
time.sleep(2)
# Click on Advance search
adv_search = driver.find_element(By.LINK_TEXT, 'Advanced Search')
time.sleep(1)
adv_search.click()

# Advance title search
adv_title_search = driver.find_element(By.LINK_TEXT, 'Advanced Title Search')
adv_title_search.click()

# Title Type
title_type1 = driver.find_element(By.ID, 'title_type-1')
title_type1.click()

title_type2 = driver.find_element(By.ID, 'title_type-2')
title_type2.click()

# Release Date Min
date = driver.find_element(By.NAME, 'release_date-min')
date.click()
date.send_keys('1990')

# Release Date Max
date = driver.find_element(By.NAME, 'release_date-max')
date.click()
date.send_keys('2020')

# User Rating Min
min_rating = driver.find_element(By.NAME, 'user_rating-min')
min_rating.click()
dropdown_1 = Select(min_rating)
dropdown_1.select_by_visible_text('1.0')

# User Rating Max
max_rating = driver.find_element(By.NAME, 'user_rating-max')
max_rating.click()
dropdown_2 = Select(max_rating)
dropdown_2.select_by_visible_text('10')

# Title Groups
title_group = driver.find_element(By.ID, 'groups-7')
title_group.click()

# Language
language = driver.find_element(By.NAME, 'languages')
dropdown_3 = Select(language)
dropdown_3.select_by_visible_text('English')

# Display Options
per_page = driver.find_element(By.ID, 'search-count')
per_page.click()
dropdown_4 = Select(per_page)
dropdown_4.select_by_index(2)

# click on the search
search_button = driver.find_element(By.XPATH, '(//button[@type="submit"])[2]')
search_button.click()

with open('imdb.csv', 'w', newline='') as f:
    the_writer= writer(f)
    header = ['Movie Title', 'Duration', 'Rating', 'Release Year', 'Genre']
    the_writer.writerow(header)


    while True:
        current_url = driver.current_url
        response = requests.get(current_url)
        soup = BeautifulSoup(response.content, 'lxml')

        # movies
        movies = soup.find_all('div', class_='lister-item mode-advanced')

        # list
        for result in movies:
            try:
                movie_name = result.find('h3', class_='lister-item-header').a.get_text() 
            except Exception:
                movie_name =[None]
            try:
                duration = result.find('span', class_='runtime').get_text() 
            except Exception:
                duration =[None]
            try:
                genre = result.find('span', class_='genre').get_text().strip() 
            except Exception:
                genre =[None]
            try:
                rating = result.find('div', class_='ratings-imdb-rating').get_text().strip() 
            except Exception:
                rating =[None]
            try:
                year = result.find('span', class_='lister-item-year').get_text().replace('(', '').replace(')', '')
            except Exception:
                year =[None]
            info = [movie_name, duration, rating, year, genre]
            the_writer.writerow(info)
        try:
            next_button = driver.find_element_by_class_name('next-page')
            next_button.click()
        except NoSuchElementException:
            break


        print(current_url)
```
